# Employee Data Analysis

This notebook explores employee data to identify patterns in salary, experience, and data quality.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Data Loading

```python
df = pd.read_csv("../data/processed/employees_cleaned.csv")
df.head()
```

## Data Overview

Understanding the structure and quality of the dataset.

```python
df.info()
df.describe()
```

## Salary by Gender

Analyzing salary differences between genders.

```python
salary_gender = df.groupby("gender")["salary"].mean()

salary_gender.plot(kind="bar")
plt.title("Average Salary by Gender")
plt.ylabel("Salary")
plt.xticks(rotation=0)

plt.savefig("../outputs/figures/salary_by_gender.png")
plt.show()
```

## Salary by Department

Comparing salary distribution across departments.

```python
salary_dept = df.groupby("department")["salary"].mean().sort_values()

salary_dept.plot(kind="barh")
plt.title("Average Salary by Department")
plt.xlabel("Salary")

plt.savefig("../outputs/figures/salary_by_department.png")
plt.show()
```

**Insight:**  
Some departments offer higher average salaries, indicating role-based compensation differences.

## Experience vs Salary

Exploring the relationship between experience and salary.

```python
plt.scatter(df["years_of_experience"], df["salary"])

plt.title("Experience vs Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")

plt.savefig("../outputs/figures/experience_vs_salary.png")
plt.show()
```

**Insight:**  
There is a strong positive correlation between experience and salary.

## Top Earners

Identifying highest-paid employees.

```python
top_earners = df.sort_values(by="salary", ascending=False).head(10)
top_earners.to_csv("../outputs/reports/top_earners.csv", index=False)

top_earners
```

## Data Quality - Phone Errors

Analyzing issues in phone number data.

```python
phone_errors = df["phone_error"].value_counts()
phone_errors.to_csv("../outputs/reports/phone_errors.csv")

phone_errors
```

**Insight:**  
A significant number of phone numbers are invalid, indicating data quality issues.
